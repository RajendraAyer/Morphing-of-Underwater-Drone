```python
#PREPARRING DATASET FOR ANALYSIS
import numpy as np
dataset = np.genfromtxt('Morph_Shapes.csv', delimiter=',', skip_header=True, usecols=(0,1,2,3))
```

```python
#MODEL DEVELOPMENT AND TRAINING
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(1, input_dim = 3, activation='linear'))

model.compile(loss = 'mse', optimizer='rmsprop', metrics=['mse'])
model.fit(x=dataset[:, 0:3], y=dataset[:,3], epochs=1060, verbose=1)
```

```python
#PREDICTIONS FROM TRAINED MODEL
input_condtions=dataset[51:52, :3]
predict=model.predict(input_condtions)
Predict=([1 if y>=0.80 else 0 for y in predict] and [2 if y>=1.80 else 1 for y in predict] and [3 if y>=2.80 else 2 for y in predict] and [4 if y>=3.80 else 3 for y in predict])
print("MORPHING OF UNDERWATER DRONE SHAPES AS PER DIFFERENT OCEAN CONDITIONS \n")
print("Different Morphing Shapes as per ocean conditions are:\n", "DUCK, PENGUIN, SEAL, SHARK \n")
print("INPUT OCEAN CONDITIONS VALUES:\n"" Depth(KM)     Pressure(BAR)     Temperature(F) \n",input_condtions)
print("\n","MORPHED SHAPE OF DRONE")
if Predict==[1]:
  print("DUCK")
elif Predict==[2]:
  print("PENGUIN")
elif Predict==[3]:
  print("SEAL")
else:
  print("SHARK")
```
